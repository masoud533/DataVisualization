```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
from scipy import stats
from sqlalchemy import create_engine
from sqlalchemy import URL
from sqlalchemy import text
from urllib.parse import quote_plus
```

```python
url_object = URL.create(
    "mysql+mysqlconnector",
    username="root",
    password="",
    host="localhost"
    )
engin = create_engine(url_object)
```

```python
query = ("SELECT * FROM EPA_Fuel_Economy;") 
```

```python
with engin.connect() as conn:
    conn.execute(text("USE dataVisualization"))
    data = pd.read_sql(text(query), conn)
```

```python
xmin, xmax = 0, 40
x = np.linspace(xmin, xmax, 100)
```

# difference between fuel parameters

```python
fig, ax = plt.subplots(figsize=(7,5))
highway = stats.norm.pdf(x, np.mean(data['highway08']),  np.std(data['highway08']))
city = stats.norm.pdf(x, np.mean(data['city08']), np.std(data['city08']))
ax.plot(x, highway, 'b', label = 'highway')
ax.plot(x, city, 'r', label = 'city') 
ax.axvline(np.mean(data['highway08']), color = 'b', linestyle= '--' )
ax.axvline(np.mean(data['city08']), color = 'r', linestyle='--')
ax.legend()
ax.set_xlabel('CarConsumption')

difference_value = round((np.mean(data['highway08']) - np.mean(data['city08'])),3)
fig.text(0.15,0.75,('Difference between average of\nreal value and estimated value\nis '+str(difference_value)))
```
